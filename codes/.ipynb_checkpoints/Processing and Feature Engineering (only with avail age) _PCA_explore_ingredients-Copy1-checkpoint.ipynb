### Calorie, Fat, Protein, Carb preference

```python
cal = df.groupby('item_id')['calories'].mean()
plt.figure(figsize=(15,5))
sns.distplot(cal, bins=40)
plt.xticks(np.arange(min(cal), max(cal)+1, 50))
plt.show()
```

```python
#Low calorie == ≤100g (≤ 40 kilocalorie per 100g for 1 serving 250g)
#High calorie == ≥750g (≥ 300 kilocalorie per 100g for 1 serving 250g)
def calorie_pref(x):
    if x <= 100 :
        return 1 #low calorie
    elif x >= 750:
        return 3 #High calorie
    else:
        return 2 

df['calorie_pref'] = df.calories.map(lambda x : calorie_pref(x))
```

```python
fat = df.groupby('item_id')['fat'].mean()
plt.figure(figsize=(15,5))
sns.distplot(fat, bins=40)
plt.xticks(np.arange(min(fat), max(fat)+1, 5))
plt.show()
```

```python
#Low fat == ≤7.5g (≤3g fat per 100g for 1 serving 250g) 
def fat_pref(x):
    if x <= 7.5 :
        return 1 #low fat
    else:
        return 0  
    
df['fat_pref'] = df.fat.map(lambda x : fat_pref(x))
```

```python
pro = df.groupby('item_id')['protein'].mean()
plt.figure(figsize=(15,5))
sns.distplot(pro, bins=40)
plt.xticks(np.arange(min(pro), max(pro)+1, 5))
plt.show()
```

```python
#Rich in protein == ≥25 (≥10g of protein per 100g for 1 serving 250g)
def pro_pref(x):
    if x >= 25 :
        return 1 #high protein
    else:
        return 0  
    
df['pro_pref'] = df.protein.map(lambda x : pro_pref(x))
```

```python
carbs = df.groupby('item_id')['carbs'].mean()
plt.figure(figsize=(15,5))
sns.distplot(carbs, bins=40)
plt.xticks(np.arange(min(carbs), max(carbs)+1, 5))
plt.show()
```

```python
#Low carbs == ≤25 (≤10 g carbohydrate per 100g for 1 serving 250g)
def carb_pref(x):
    if x <= 25 :
        return 1 #low carvs
    else:
        return 0  
    
df['carb_pref'] = df.carbs.map(lambda x : carb_pref(x))
```

## Modelling

```python
df_merge = pd.concat([df,df_cat,df_mw], axis=1)
```

```python
df_merge_mean = df_merge.groupby('user_id').mean()
```

```python
df_merge_mean.sample(5)
```

```python
df[df.user_id == 40025][['calorie_pref', 'fat_pref','pro_pref', 'carb_pref','is_chilled',]].mean()
```

```python
df_merge_mean['calorie_pref_norm'] = df_merge_mean.calorie_pref/1.729730
df_merge_mean['fat_pref_norm'] = df_merge_mean.fat_pref/0.324324
df_merge_mean['pro_pref_norm'] = df_merge_mean.pro_pref/0.540541
df_merge_mean['carb_pref_norm'] = df_merge_mean.carb_pref/0.324324
df_merge_mean['is_chilled_norm'] = df_merge_mean.is_chilled/0.108108
```

```python
features = ['region','use_wallet','has_coupon',
            'age','first', 'second','third',
            'delivery_fee','gave_feedback','groupbuyer',
            'unit_price', 'drink', 'favorite', 
            'guilt-free', 'highlight', 'local', 
            'main', 'off_menu', 'regular',
            'side','calorie_pref_norm', 'fat_pref_norm', 
            'pro_pref_norm', 'carb_pref_norm','is_chilled_norm'
           ]
```

```python
X = df_merge_mean[features].round(2)
```

```python
X.sample(5)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(X.corr(), annot=True, cmap='RdBu_r')
plt.show()
```

```python
ss = StandardScaler()
Xs = ss.fit_transform(X)
```

### PCA

```python
pca = PCA()
principalComponents = pca.fit_transform(Xs)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_)
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)

# Save components to a DataFrame
PCA_components = pd.DataFrame(principalComponents)
```

```python
plt.scatter(PCA_components[0], PCA_components[1], alpha=.1, color='black')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

### KMeans

```python
ks = range(1, 10)
inertias = []
for k in ks:
    # Create a KMeans instance with k clusters: model
    knn_model = KMeans(n_clusters=k)
    
    # Fit model to samples
    knn_model.fit(PCA_components.iloc[:,:2])
    
    # Append the inertia to the list of inertias
    inertias.append(knn_model.inertia_)
    
plt.plot(ks, inertias, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
model = KMeans(n_clusters=2)
model.fit(PCA_components.iloc[:,:2])
```

```python
pred = model.labels_
```

```python
X['cluster'] = pred
```
